### Importing

```python
import pandas as pd
import openpyxl as xl
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score
```

### Reviews

```python
class Reviews:
    def __init__(self,text,score):
        self.text=text
        self.score=score
        self.sentiment=self.sentiment()
    def sentiment(self):
        if self.score <= 2:
            return "Negative"
        elif self.score == 3:
            return "Neutral"
        else:
            return "Positive"
        
```

#### Gathering data

```python
wb=xl.load_workbook('amazon_data.xlsx')
sheet=wb['amazon_data']
data=[]
for i in range(2,sheet.max_row):
    obtained1=sheet.cell(i,6)
    obtained2=sheet.cell(i,7)
    data.append(Reviews(obtained1.value,obtained2.value))
```

### Preparing data

```python
training,testing = train_test_split(data,test_size=0.30,random_state=33)
train_x = [x.text for x in training]
train_y = [x.sentiment for x in training]

test_x = [x.text for x in testing]
test_y = [x.sentiment for x in testing]

```

###### Here the text has been Converted to integers because machine learning modules prefers numbers over text

```python
vector=CountVectorizer()
train_x_vector = vector.fit_transform(train_x)
test_x_vector = vector.transform(test_x)
```

## Classifier

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(train_x_vector,train_y)
predictions=model.predict(test_x_vector)
print(predictions)
score=accuracy_score(predictions,test_y)
print(score)
```

```python
new=["i liked the book","i didn't liked the book","very bad book complete waste of time"]
new_vector=vector.transform(new)
prediction_test=model.predict(new_vector)
print(prediction_test)
```
